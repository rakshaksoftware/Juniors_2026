```python

import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y)
plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
x = np.random.uniform(0,1,100)
y = np.random.normal(0.5,0.1,100)

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show()
data = np.random.exponential(1, 500)

plt.hist(data, bins=30)
plt.title("Exponential Distribution Histogram")
plt.show()
```

```python
import pandas as pd

data = {
    "Name": ["Alice","Bob","Charlie","Diana","Ethan"],
    "Age": [28,34,25,42,30],
    "Salary": [70000,80000,50000,110000,75000]
}

df = pd.DataFrame(data)

print("Original DataFrame:\n", df)
df["Tax"] = df["Salary"] * 0.2
print(df[df["Age"] >= 30])
avg_salary = df[df["Age"] < 30]["Salary"].mean()
print("Average Salary (<30):", avg_salary)
print(df.sort_values(by="Salary", ascending=False))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dates = pd.date_range("2025-01-01","2025-01-31")

days = np.arange(len(dates))

temperature = 15 + 10*np.sin(2*np.pi*days/31)

temp_series = pd.Series(temperature, index=dates)
plt.plot(temp_series)
plt.title("Daily Temperature - Jan 2025")
plt.ylabel("Temperature (°C)")
plt.show()
rolling_avg = temp_series.rolling(window=7).mean()

plt.plot(temp_series, label="Temperature")
plt.plot(rolling_avg, label="7-Day Rolling Avg")

plt.legend()
plt.show()
print(temp_series[temp_series > 20])
```
