```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
#line plot :
x = np.linspace( -10 , 10 , 200 ) 

print(x) 

y = x**2 

print( y ) 

plt.plot( x , y ) 
plt.show()
```

```python
# scatter plot 

x1 = np.linspace( 0 , 1 , 100 )
y1 = np.random.normal( loc = 0.5 , scale = 0.1 , size = 100 )

plt.scatter( x1 , y1 )

plt.show() 
```

```python
fig, (ax1 , ax2) = plt.subplots( 1 , 2 ) 

ax1.plot( x , y )
ax1.set_title( "line plot" )

ax2.scatter( x1 , y1 )
ax2.set_title( "scatter plot" )

plt.show()
```

```python
np.random.seed( 43 )

data = np.random.exponential( scale = 1 , size = 500 ) 
print(data) 

plt.hist( data , bins = 20 , color = 'red' , edgecolor = 'black' )

plt.show() 
```
