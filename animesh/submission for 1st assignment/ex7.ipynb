```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
x = np.arange(1,32)

dates = pd.date_range( start= '2026-01-01' , periods= 31 )

y = 15 + 10*np.sin( (2*np.pi*x) / len(x) ) 

df = pd.DataFrame({'Date' : dates,
                  'Tempreature' : y})

plt.plot( df['Date'] , df['Tempreature'] , marker = 'o') 

plt.title( 'Temp vs Dates for January 2025' )
plt.xlabel( 'Dates' )
plt.ylabel( 'Temp ( in deg celcius )')
plt.xticks(df['Date'][::5])
plt.xticks(rotation= 45)

plt.grid(True)

plt.show()
```

```python
df['Rolling Avg'] = df['Tempreature'].rolling(window=7).mean()

plt.plot( df['Date'] , df['Rolling Avg'] , marker = 'o' , color = 'red' )
plt.xticks(df['Date'][::5])
plt.xticks(rotation= 45)

plt.grid(True)


plt.show()
```

```python

plt.plot( df['Date'] , df['Tempreature'] , marker = 'o') 
plt.plot( df['Date'] , df['Rolling Avg'] , marker = 'o' , color = 'red' )

plt.title( 'Temp vs Dates for January 2025' )
plt.xlabel( 'Dates' )
plt.ylabel( 'Temp ( in deg celcius )')
plt.xticks(df['Date'][::5])
plt.xticks(rotation= 45)

plt.show()

```
