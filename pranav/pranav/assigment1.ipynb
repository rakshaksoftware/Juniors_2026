```python
for i in range(1,101): 
    if(i%15==0):
        print("FizzBuzz")
        continue
    if(i%3==0):
        print("Fizz")
        continue
    if(i%5==0):
        print("Buzz")
        continue
    print(i)
```

```python
# Exercise 1.2: Prime numbers within a range 
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

for i in range(a, b + 1):
    if i > 1:
        for j in range(2, i):
            if (i % j) == 0:
                break 
        else: 
            print(i)
```

```python
#excercise 2.1
A= np.random.randint(0, 10, size=(3, 3))
B= np.random.randint(0, 10, size=(3, 3))
print(A)
print(B)
print(A+B)
print(np.multiply(A,B))
print(np.dot(A,B))
print(np.linalg.det(A))
```

```python
#excercise 2.2
array=np.random.normal(loc=50,scale=5,size=1000)
print(np.mean(array))  
print(np.median(array))
print(np.std(array))
print(np.var(array))
print(np.percentile(array, 25))
print(np.percentile(array, 75))
```
